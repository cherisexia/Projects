## Creating Stored Procedures

### Create a stored procedure to update the total invoice amount for a specific customer 

```python
%%sql
drop procedure if exists UpdateCustomerTotal;

CREATE PROCEDURE UpdateCustomerTotal(IN customerID INT, IN newTotal DOUBLE)
BEGIN
  #-- Check if the customer exists and has an invoice
  IF EXISTS(SELECT 1 FROM sample.invoice WHERE CustomerId = customerID) THEN #select 1 is a placeholder in this case
    
    #-- Check if the new total is greater than a certain amount
    IF newTotal > 10 THEN
      #-- If so, give a 10% discount and update the sample.invoice
      UPDATE sample.invoice
      SET Total = newTotal * 0.9
      WHERE CustomerId = customerID AND Total < newTotal;
      
      #-- You can select the updated rows to get the updated totals
      SELECT invoiceId, customerId, Total
      FROM sample.invoice
      WHERE CustomerId = customerID AND Total = newTotal * 0.9;
    
    ELSE
      UPDATE sample.invoice
      SET Total = newTotal
      WHERE CustomerId = customerID AND Total < newTotal;
      
      #-- Select the updated rows to get the updated totals
      SELECT InvoiceId, Total
      FROM sample.invoice
      WHERE CustomerId = customerID AND Total = newTotal;
    END IF;
    
  end if;

END;
```

### Execute the stored procedure

```python
%%sql
call UpdateCustomerTotal(2, 12) #check the results
```

### Create a transaction to perform group operations on the Invoice table

```python
%%sql

START TRANSACTION;

INSERT INTO sample.invoice (invoiceId, CustomerId, BillingCountry, total) VALUES (4, 10, 'China', 11.3);

SAVEPOINT AFTER_INSERT;

#-- Apply discount if needed
UPDATE sample.invoice SET total = total * 0.9 WHERE customerid = 10;

#-- Assume we encounter a problem here and decide to roll back to the savepoint
ROLLBACK TO SAVEPOINT AFTER_INSERT;

UPDATE sample.invoice SET invoicedate = '2024-01-01' 
WHERE invoiceId = 4;

COMMIT;
```

## Create a trigger to keep track of the changes in the total invoice amount

```python
%%sql
Create table if not exists Sample.total_changes(
    change_id INT AUTO_INCREMENT PRIMARY KEY,
    customerid INT,
    old_total double,
    new_total double,
    change_date TIMESTAMP DEFAULT CURRENT_TIMESTAMP
    
)
```

```python
%%sql
use Sample;
delimiter //
CREATE TRIGGER before_total_update BEFORE UPDATE ON sample.invoice
FOR EACH ROW
BEGIN
    IF OLD.total != NEW.total THEN
        INSERT INTO sample.total_changes (customerid, old_total, new_total) 
        VALUES (OLD.customerid, OLD.total, NEW.total);
    END IF;
END;//
delimiter ;
```

## Visulize the total invoice amount by categories

```python
import matplotlib.pyplot as plt
import sqlalchemy
import pandas as pd

# Replace these with your actual database connection details
username = 'root'
password = 'xyh001205'
hostname = 'localhost'
database_name = 'mysql'

# Create an engine to connect to the MySQL database
# Note: You might need to adjust the URL based on your database and connector
engine = sqlalchemy.create_engine(f'mysql+pymysql://{username}:{password}@{hostname}/{database_name}')

sql_query = """
WITH temp AS (
  SELECT
    CASE 
      WHEN total > 0 AND total < 5 THEN 'Total between 0-5'
      WHEN total >= 5 AND total < 10 THEN 'Total between 5-10'
      WHEN total >= 10 THEN 'Total >= 10'
    END AS TotalAmountCategory
  FROM Sample.Invoice
)

SELECT
  SUM(CASE WHEN TotalAmountCategory = 'Total between 0-5' THEN 1 ELSE 0 END) AS Total_between_0_5,
  SUM(CASE WHEN TotalAmountCategory = 'Total between 5-10' THEN 1 ELSE 0 END) AS Total_between_5_10,
  SUM(CASE WHEN TotalAmountCategory = 'Total >= 10' THEN 1 ELSE 0 END) AS Total_greater_than_10
FROM temp;

"""

# Execute the query and load the results into a pandas DataFrame
df = pd.read_sql_query(sql_query, engine)

print(df)

# Plotting the results
df.plot(kind='bar')
plt.ylabel('Counts')
plt.xlabel('Category')

plt.title('Invoice amount by categories')
#plt.xticks(rotation=45)
plt.show()
```
